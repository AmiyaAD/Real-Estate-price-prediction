```python
rm = RandomForestRegressor(n_estimators=200,criterion='squared_error', min_samples_split=4,
    min_samples_leaf=2,
    min_weight_fraction_leaf=0.0,
    max_features=1.0,)
rm.fit(x_train,y_train)
rm.score(x_train,y_train)*100, rm.score(x_test,y_test)*100
```

```python
mse5 = mean_absolute_error(y_test, rm.predict(x_test))
print(mse5)
rmse5 = np.sqrt(mse5)
print(rmse5)
```

```python
# x_test
```

```python
pipeline = Pipeline ([
    ("si", SimpleImputer(strategy="mean")),
    ("ss", StandardScaler()),
    ("rm", RandomForestRegressor(n_estimators=200,criterion='squared_error', min_samples_split=4,
    min_samples_leaf=2,
    min_weight_fraction_leaf=0.0,
    max_features=1.0))
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
# x_test
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(rm,x_test,y_test,scoring="neg_mean_squared_error",cv=10)
rms_score = np.sqrt(-scores)
```

```python
rms_score
```

```python
def print_scores(scores):
    print("Score", scores),
    print("Mean", scores.mean())
    print("Std", scores.std())
print_scores(rms_score)    
```
